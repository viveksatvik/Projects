```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

from PIL import Image
from wordcloud import WordCloud, ImageColorGenerator
from wordcloud import WordCloud
```

```python
''' reading data '''
df = pd.read_csv("../input/game-of-thronesgot/game of thrones.csv")

''' displaying first five rows of data '''
df.head()
```

```python
''' joining all titles '''
t1 = " ".join(title for title in df.Title)
```

```python
''' creaing word cloud'''
wc1 = WordCloud(collocations=False, background_color='white', width=3840, height=2160).generate(t1)

''' svaing an image'''
wc1.to_file('got.png')
```

```python
''' Displaying an image '''
plt.imshow(wc1, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
''' rading file '''
t2 = open('../input/poetry/disney.txt', 'r').read()
```

```python
''' converting image to an array '''
backg_img = np.array(Image.open('../input/disney-characters-dataset/cartoon/test/mickey/28de32146d9bfb991d0e859fca65f012.jpg'))

''' plotting an image'''
plt.imshow(backg_img)
```

```python
''' creating 2nd word cloud for second file t2 '''
wc2 = WordCloud(background_color='white', mask=backg_img,  width=3840, height=2160).generate(t2)

''' font color matching the masked imag'''
img_c = ImageColorGenerator(backg_img)
wc2.recolor(color_func=img_c)

''' saving the image '''
wc2.to_file('disney.png')
```

```python
''' plotting wc2 '''
plt.imshow(wc2, interpolation='bilinear')
plt.axis("off")
plt.show()
```
